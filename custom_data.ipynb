```python
import numpy as np
import matplotlib.pyplot as plt

# a = np.random.randn(3000)
a = np.random.randn(3000)*0.1
cp1 = 800
cp2 = 1400
a[:cp1] += (np.arange(1,cp1+1,1)*0.001)
a[cp1:cp2] += np.sin(1/48*np.arange(1,cp2-cp1+1,1))+a[cp1-1]
a[cp2:] += -np.arange(1,len(a)-cp2+1,1)*0.005 + a[cp2-1]

a = np.array(a)
np.save('data/sim/simdata1.npy',a)

plt.plot(np.load('data/sim/simdata1.npy'))
```

```python
b = np.random.randn(3000)*0.1
cp1 = 400
cp2 = 2100
b[:cp1] += -(np.arange(1,cp1+1,1)*0.003)
b[cp1:cp2] += np.sin(1/23*np.arange(1,cp2-cp1+1,1))*0.74+b[cp1-1]
b[cp2:] += np.arange(1,len(b)-cp2+1,1)*0.005 + b[cp2-1]

b = np.array(b)
np.save('data/sim/simdata2.npy',b)
plt.plot(np.load('data/sim/simdata2.npy'))
```
